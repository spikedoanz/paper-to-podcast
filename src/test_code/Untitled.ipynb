```python
from bark import SAMPLE_RATE, generate_audio, preload_models
from scipy.io.wavfile import write as write_wav
from IPython.display import Audio
import numpy as np
import json
import re
# download and load all models
preload_models()
# generate audio from text

with open ("/home/tama/work/pdf-podcast/examples/16:22-17-08-2023.json", "r") as f:
    data = json.load(f)['record']
audio_array = []
# save audio to disk
for i in range(len(data)):
    text = data[i]['content']
    role = data[i]['role']
    if role == 'user':
        voice = "v2/en_speaker_1"
    else:
        voice = "v2/en_speaker_6"
    sentences = re.split(r'[.!?]', text)
    sentences = [sentence.strip() for sentence in sentences if sentence.strip()]
    for _ in sentences:
        audio_array.extend(generate_audio(_, history_prompt=voice))
    print(text)
```

```python
savepath = "./recording/"
filename = 'test.wav'
savefile = np.array(audio_array)
write_wav(savepath + filename, sample_rate, savefile)
  
```
